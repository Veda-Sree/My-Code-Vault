```python
import random

def number_guessing_game():
    print("Welcome to the Number Guessing Game!")
    print("I'm thinking of a number between 1 and 100. Try to guess it!")
    number = random.randint(1, 100)
    attempts = 0
    while True:
        guess = int(input("Enter your guess: "))
        attempts += 1
        if guess < number:
            print("Too low! Try guessing a higher number.")
        elif guess > number:
            print("Too high! Try guessing a lower number.")
        else:
            print(f"Congratulations! You've guessed the number {number} in {attempts} attempts.")
            break

def rock_paper_scissors():
    print("Welcome to Rock, Paper, Scissors!")
    choices = ['rock', 'paper', 'scissors']
    user_choice = input("Enter your choice (rock, paper, scissors): ").lower()
    computer_choice = random.choice(choices)
    print(f"Computer chooses: {computer_choice}")
    if user_choice == computer_choice:
        print("It's a tie!")
    elif (user_choice == 'rock' and computer_choice == 'scissors') or \
         (user_choice == 'paper' and computer_choice == 'rock') or \
         (user_choice == 'scissors' and computer_choice == 'paper'):
        print("Congratulations! You win!")
        return 1
    else:
        print("Computer wins! Better luck next time.")
        return 0

def hangman():
    print("Welcome to Hangman!")
    words = ["apple", "banana", "cat", "dog", "elephant", "fish", "giraffe", "horse", "ice cream", "jaguar", "kangaroo", "lion", "monkey", "narwhal", "octopus", "penguin", "quail", "rabbit", "snake", "tiger", "umbrella", "volcano", "whale", "xylophone", "yak", "zebra"]
    word = random.choice(words)
    print(f"The word to guess has {len(word)} letters.")
    guessed_letters = []
    tries = 6
    while tries > 0:
        word_display = "".join([letter if letter in guessed_letters else "_" for letter in word])
        print(f"Word: {word_display}")
        if "_" not in word_display:
            print("Congratulations! You've guessed the word correctly.")
            break
        guess = input("Guess a letter: ").lower()
        if guess in guessed_letters:
            print("You've already guessed this letter. Try again.")
        elif guess in word:
            print("Correct guess!")
            guessed_letters.append(guess)
        else:
            print("Incorrect guess!")
            tries -= 1
            print(f"Tries left: {tries}")
    else:
        print(f"Out of tries! The word was '{word}'.")

def dice_rolling_game():
    print("Welcome to Dice Rolling Game!")
    print("I'm going to roll two dice. Try to guess the total.")
    total = random.randint(2, 12)
    guess = int(input("Enter your guess (between 2 and 12): "))
    if guess == total:
        print(f"Congratulations! You've guessed the total {total} correctly.")
    else:
        print(f"Sorry, the correct total was {total}.")

def main():
    games = {
        "1": ("Number Guessing Game", number_guessing_game),
        "2": ("Rock, Paper, Scissors", rock_paper_scissors),
        "3": ("Hangman", hangman),
        "4": ("Dice Rolling Game", dice_rolling_game)
    }
    while True:
        print("\nChoose a game to play:")
        for key, (name, _) in games.items():
            print(f"{key}. {name}")
        print("5. Quit")
        choice = input("Enter your choice: ")

        if choice == '5':
            print("Thanks for playing! Goodbye.")
            break

        game = games.get(choice)
        if game:
            print(f"\nLet's play {game[0]}!")
            while True:
                game[1]()
                play_again = input("Do you want to play the same game again? (yes/no): ").lower()
                if play_again != 'yes':
                    break
        else:
            print("Invalid choice. Please enter a number between 1 and 5.")

if __name__ == "__main__":
    main()
```
